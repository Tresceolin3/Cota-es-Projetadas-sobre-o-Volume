```python
import talib
from talib.abstract import *
from talib import MA_Type
```

```python
# imprime informações sobre uma função em específico
print(CDLHAMMER)
```

```python
print(CDLHAMMER.info)
```

```python
print(CDLHAMMER.input_names)
```

```python
talib.CDLHAMMER(df.open, df.high, df.low, df.close).value_counts()
```

```python
is_hammer = talib.CDLHAMMER(df.open, df.high, df.low, df.close) == 100
```

```python
df.loc[is_hammer]
```
