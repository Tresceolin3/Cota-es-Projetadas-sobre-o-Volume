```python
def projeta_volume(df, k):
    i = 0
    while i < len(df):
        volume = 0
        j = i
        while volume < k and j < len(df):
            volume += df['volume'][j]
            j += 1
        # Agrupar os vários candles em um só
        novo_candle = {
            'date': df.index[i],
            'open': df['open'][i],
            'high': max(df['high'][i:j]),
            'low': min(df['low'][i:j]),
            'close': df['close'][j-1],
            'volume': volume
        }
        i = j
        break
```

```python
for i in range(len(df) - 1):
    df.loc[df.index[i], '1_ahead'] = df.iloc[i+1].close > df.iloc[i].open
```

```python
df['2_ahead'] = 0    
for i in range(len(df) - 2):
    df.loc[df.index[i], '2_ahead'] = df.iloc[i+2].close > df.iloc[i].open
```

```python
df['3_ahead'] = 0  
for i in range(len(df) - 3):
    df.loc[df.index[i], '3_ahead'] = df.iloc[i+3].close > df.iloc[i].open
```

```python
df['4_ahead'] = 0  
for i in range(len(df) - 4):
    df.loc[df.index[i], '4_ahead'] = df.iloc[i+4].close > df.iloc[i].open
```

```python
df['5_ahead'] = 0  
for i in range(len(df) - 5):
    df.loc[df.index[i], '5_ahead'] = df.iloc[i+5].close > df.iloc[i].open
```

```python
from scipy.stats import pearsonr


ahead_candle = patterns + ['1_ahead', '2_ahead', '3_ahead', '4_ahead', '5_ahead']
def_person = df[ahead_candle]


correlation_results = {}
for pattern in patterns:
    correlation, p_value = pearsonr(def_person[pattern], def_person['2_ahead'])
    correlation_results[pattern] = correlation


for pattern, correlation in correlation_results.items():
    print(f"Correlation between {pattern} and '2_ahead': {correlation}")
```

```python
import pandas as pd
from scipy.stats import pearsonr


ahead_columns = ['1_ahead', '2_ahead', '3_ahead', '4_ahead', '5_ahead']

correlation_results = {}


for pattern in patterns:
  
    pattern_correlations = []
    

    for column in ahead_columns:
       
        correlation, _ = pearsonr(df[pattern], df[column])
        
       
        pattern_correlations.append(correlation)
    
    
    correlation_results[pattern] = pattern_correlations


for pattern, correlations in correlation_results.items():
    print(f"Correlations for {pattern}:")
    for i, column in enumerate(ahead_columns):
        print(f"{column}: {correlations[i]}")
    print()
```
