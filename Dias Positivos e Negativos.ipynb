```python
import pandas as pd
import matplotlib.pyplot as plt

arquivos_csv = [
     'ABEV3.Semanal.csv',
     'B3SA3.Semanal.csv',
     'BBAS3.Semanal.csv',
     'BBDC4.Semanal.csv',
     'ELET3.Semanal.csv',
     'ITSA4.Semanal.csv',
     'ITUB4.Semanal.csv',
     'PETR3.Semanal.csv',
     'PETR4.Semanal.csv',
     'VALE3.Semanal.csv',
     'WEGE3.Semanal.csv'
]

fig, ax = plt.subplots()

for arquivo in arquivos_csv:
    df = pd.read_csv(arquivo)
    symbol = arquivo.split('.')[0]

    dias_p = df[df['Close'] >= df['Open']].shape[0]
    dias_n = df[df['Close'] < df['Open']].shape[0]
    dias_total = df['Date'].count()

    
    p_positivo = dias_p / dias_total * 100
    p_negativa = dias_n / dias_total * 100

    
    color_positive = 'green'
    color_negative = 'red'

   
    ax.bar(symbol, p_positivo, color=color_positive)
    ax.bar(symbol, p_negativa, bottom=p_positivo, color=color_negative)


ax.set_xlabel('Dados Semanais')
ax.set_ylabel('Porcentagem de dias')
ax.set_title('Dias Positivos e Negativos')


plt.xticks(rotation=90)


ax.legend(['Positivos', 'Negativo'], bbox_to_anchor=(1, 1), loc='upper left')

# Exibe o gráfico
plt.show()


```
